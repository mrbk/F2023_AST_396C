## Q1 

(a) The density is constant, $\rho(t)=\rho(t_0)=\rho_0$, and the universe is given to be flat ($k=0$). So, we can write the Friedman equation as $$H(a)^2=\left(\frac{\dot{a}}{a}\right)^2=\frac{8\,\pi\,G}{3}\,\rho_0\;.$$ We therefore have $$\frac{d\ln a}{dt}=\sqrt{\frac{8\,\pi\,G}{3}\,\rho_0}\,,$$ with the solution $$a(t)=e^{H_0(t-t_0)}$$ (recall that $H_0^2=8\,\pi\,G\,\rho_0/3$ in this case because $k=0$ and therefore $\rho(t_0)=\rho_{\rm c,0}$). Apparently, the scale factor changes exponentially with time in this steady-state Universe.

(b) The fluid equation is $$\dot{\rho}=-3 H(a)\,\rho\,(1+w)\,.$$ From the Friedmann equation in part (a), we have that $H(a)=H_0$, and it's given that $\rho(a)=\rho(a=1)=\rho(t_0)$. As a result, $\dot{\rho}=0$, and since $H(a) \ne 0$, we must have that $$w=-1\,.$$ This is exactly the same as the equation of state for dark energy ($\Lambda$). In retrospect, perhaps this is unsurprising, as dark energy also has the property that $\rho(a)=\rho(a=1)=\rho(t_0)$.

(c) The idea here is that the density is constant, but since the universe is expanding, we must have that mass is being created because the mass present is constantly being diluted by the expansion. We can express this as $$\dot{\rho}=0 \leftrightarrow \frac{d(M/V)}{dt}=0\,.$$ Normally, if we look at the mass contained within a comoving volume $V_{\rm c}=a^{-3}V$, it is constant, but in this case, that isn't the case. So, we have that
\begin{equation*}
\frac{d(M/V)}{dt}=-M\,\frac{\dot{V}}{V^2}+\frac{\dot{M}}{V}\label{eq:rhodot}\tag{1}
\end{equation*}
with $$\dot{V}=d((a\,x)^3)/dt=3a^2\,\dot{a}x^3=3\frac{\dot{a}}{a}V\;.$$ Putting everything together, we have $$\dot{\rho}=0=-3\,M\frac{\dot{a}}{a}V+\dot{M}\,V$$ or $$\dot{M}=3\frac{\dot{a}}{a}\,M\;.$$ If we note that $\dot{a}/a=H=H_0$, then this equation can be written as $$\dot{M}=3\,H_0\,M\;,$$ so the required mass creation rate (per volume) is $$\frac{\dot{M}}{V}=\dot{\rho}_{+}=3\,H_0\,\rho=3\,H_0\,\rho_{0}\;,$$ where I've used $\dot{\rho}_{+}$ to indicate that this is the change in density coming from the creation of matter (this is the second term in Equation \ref{eq:rhodot}; the other term is equal in magnitude and opposite in sign so that $\dot{\rho}=0$). Since $k=0$, we can eliminate $\rho_{0}$ in favor of $H_0$: $$\dot{\rho}_{+}=\frac{9}{8\,\pi\,G}{H_0^3}\;.$$ Numerically, in MKS units this is:

```python
rate_mks=9/(8*numpy.pi*6.67e-11)*(70*1e3/3.086e22)**3
print(f'Mass creation rate per volume (MKS units): {rate_mks:.2e}.')
```

The mass creation rate is therefore $$6.27\times 10^{-44}\,{\rm kg \,s^{-1}\,m^{-3}}\,,$$ or, in $M_{\odot}\,{\rm Gyr^{-1}\,Gpc^{-3}}$:

```python
rate_astro=rate_mks/2e30*3.14e7*1e9*(3.086e25)**3
print(f'Mass creation rate per volume (astro units): {rate_astro:.2e},')
```

i.e., $2.89\times 10^{19}\,M_{\odot}\,{\rm Gyr^{-1}\,Gpc^{-3}}$. This rate is **tiny**, as most easily seen from the MKS value -- it would be roughly one proton per cubic meter per Gyr, or one proton per volume of the Sun per second. It's less intuitive in the 'Astro' units, but that's part of the point here: a Gpc is an *enormous* volume. Other measurements -- e.g., of the Universe's expansion history -- strongly rule out this steady-state theory, however.

# Q2

The Friedmann equation is $$\left(\frac{\dot{a}}{a}\right)^2=\frac{8\,\pi\,G}{3}\rho-k\frac{c^2}{a^2}\;.$$ If we multiply through by $a^2$, we get $$\dot{a}^2=\frac{8\,\pi\,G}{3}\rho\,a^2-k\,c^2\,.$$ Taking a time derivative (and remembering that $k$ is a constant) gives us $$2\,\dot{a}\ddot{a}=\frac{8\,\pi\,G}{3}\,(2\,a\,\dot{a}\rho+\dot{\rho}\,a^2)\;$$ or $$\dot{a}\ddot{a}=\frac{4\,\pi\,G}{3}\,\rho\,a^2\,\left(2\,\frac{\dot{a}}{a}+\frac{\dot{\rho}}{\rho}\right)\,.$$ We now can use the fluid equation to eliminate $\dot{\rho}$:
$$\dot{a}\ddot{a}=\frac{4\,\pi\,G}{3}\,\rho\,a^2\,\left(2\,\frac{\dot{a}}{a}-3\frac{\dot{a}}{a}(1+w)\right)\,,$$ so 
$$\dot{a}\ddot{a} =-\frac{4\,\pi\,G}{3}\,\frac{\dot{a}}{a}\,\rho\,a^2\,(1+3\,w)\,.$$ Rearranging a little, we get that $$\frac{\ddot{a}}{a}=-\frac{4\,\pi\,G}{3}\,\rho\,(1+3\,w)\,.$$ If we add back in the possibility of multiple components, i.e. $\rho=\sum_i\,\rho_{i}$, then we have that 
\begin{equation*}
\frac{\ddot{a}}{a}=-\frac{4\,\pi\,G}{3}\,\sum_i\rho_i\,(1+3\,w_i)\,.
\label{eq:acc_eq}
\tag{2}
\end{equation*}
This is the acceleration equation. Note that we got this directly from the Friedmann equation plus the fluid (continuity) equation, meaning the three equations are not independent of each other. So long as $w>-1/3$, $\ddot{a}$ will be negative (the Universe will be decelerating).

If we want to rewrite this in terms of the deceleration parameter $q_0 \equiv -\ddot{a} \, a/\dot{a}^2$, then we note that we just need to divide Equation \ref{eq:acc_eq} by $-H(a)^2=-(\dot{a}/{a})^2=-8\,\pi\,G\,\rho_{\rm c}/3$:
$$q_0=\frac{1}{2}\sum_i\Omega_i\,(1+3\,w_i)\,.$$ For the case of a flat Universe with $\Omega_{\rm m}+\Omega_{\Lambda}=1$, we therefore have that $$q_0=\frac{3}{2}\Omega_{\rm m,0}-1\,.$$ (We could also have written this as $q_0=\Omega_{\rm m,0}/2-\Omega_{\Lambda, 0}$). If $\Omega_{\rm m,0} < 2/3$, then the expansion of the Universe is accelerating in a flat $\Lambda$CDM model.

# Q3

We are interested in deriving the dependence of $\Omega_{\rm m}(a)$ on $\Omega_{\rm m,0}$ and $a$. Since we have 
\begin{equation*}
\Omega_{\rm m}(a)=\frac{\rho_{\rm m}(a)}{\rho_{\rm c}(a)}\label{eq:oma}\tag{3}\,,
\end{equation*}
and we know how $\rho_{\rm m}$ evolves with scale factor -- $\rho_{\rm m}(a)=\rho_{\rm m,0}\,a^{-3}$ -- we need to understand how $\rho_{\rm c}(a)$ evolves. But how?!? ðŸ˜«
Back to the Friedmann equation! $$\left(\frac{\dot{a}}{a}\right)^2=\frac{8\,\pi\,G}{3}\left(\rho_{\rm m}(a)+\rho_{\Lambda}(a)\right)$$ for $k=0$, so we have that $$\left(\frac{\dot{a}}{a}\right)^2=\frac{8\,\pi\,G}{3}\left(\rho_{\rm m, 0}\,a^{-3}+\rho_{\Lambda,0}\right)\,.$$ Multiplying and dividing by $\rho_{\rm c,0}$ and remembering that $H_0^2=8\,\pi\,G\,\rho_{\rm c,0}/3$, we have 
\begin{equation*}
H(a)^2=\frac{8\,\pi\,G}{3}\rho_{\rm c}(a)=H_0^2\,(\Omega_{\rm m,0}a^{-3}+\Omega_{\Lambda,0}) = \frac{8\,\pi\,G}{3}\rho_{\rm c,0}(\Omega_{\rm m,0}a^{-3}+\Omega_{\Lambda,0})\label{eq:h2}\tag{4}\;
\end{equation*}
because $H_0^2=8\,\pi\,G\,\rho_{\rm c,0}/3$. Nice! That means that we have an expression for $\rho_{\rm c}(a)$ in terms of $\Omega_{\rm m,0}$ and $\Omega_{\Lambda, 0}$ (and $a$). Going back to Eq. \ref{eq:oma}, we have 
$$\Omega_{\rm m}(a)=\frac{\rho_{\rm m,0}\,a^{-3}}{\rho_{\rm c}(a)}=\frac{\rho_{\rm m,0}\,a^{-3}}{\rho_{\rm c,0}(\Omega_{\rm m,0}a^{-3}+\Omega_{\Lambda,0})}\,.$$ We can simplify this to our final answer:
$$\Omega_{\rm m}(a)=\frac{\Omega_{\rm m,0}\,a^{-3}}{\Omega_{\rm m,0}a^{-3}+(1-\Omega_{\rm m,0})}\,.$$ Note that this is *not* the same as how $\rho_{\rm m}(a)$ evolves, which is the central point of this exercise: knowing how a govem $\rho$ evolves with $a$ is straightforward, knowing how a given $\Omega$ evolves with a is **not** because $\Omega_{i}$ depends on $\rho_{\rm c}(a)$ via Eq. \ref{eq:h2}, which means you need to know how **every** contribution to the energy density evolves to understand how $\Omega_{i}$ depends on time.

```python
# Omega values at z=0 (today):
om0=0.3
or0=9e-5
ox0=1-om0-or0
# equation of state for substance X:
wx=-3./4
```

```python
# first, let's plot rhos:
avals=numpy.logspace(-5, 0)

plt.loglog(avals, om0*avals**-3, color='k', label=r'$\rho_{\rm m}$', lw=2)
plt.loglog(avals, or0*avals**-4, color='m', label=r'$\rho_{\rm r}$', lw=2)
plt.loglog(avals, ox0*avals**(-3*(1+wx)), color='c', label=r'$\rho_{\rm X}$', lw=2)
plt.xlabel('scale factor $a$')
plt.ylabel(r'$\rho_{i}/\rho_{\rm c,0}$')
plt.legend(frameon=False)
plt.xlim([avals[0], avals[-1]])
show()
```

Now, let's plot $\{\Omega_{i}(a)\}$ values. We need $$\Omega_{i}(a)=\rho_{i}(a)/\rho_{\rm c}(a)=\frac{\rho_{i}(a)}{\rho_{\rm c,0}}\,\frac{\rho_{\rm c,0}}{\rho_{\rm c}(a)}\;,$$ which means we need to compute the second fraction (the first fraction is exactly what we plotted above). We therefore need an expression for $\rho_{\rm c}(a)/\rho_{\rm c,0}$, and fortunately, we know this is just $(H(a)/H_0)^2$:

```python
# rho_c(a)/rho_c(a=1):
def rho_c_scaled(a):
    return om0*a**(-3) + or0*a**(-4) + ox0*a**(-3*(1+wx))


plt.figure()
plt.semilogx(avals, om0*avals**-3/rho_c_scaled(avals), color='k', label=r'$\rho_{\rm m}$', lw=2)
plt.semilogx(avals, or0*avals**-4/rho_c_scaled(avals), color='m', label=r'$\rho_{\rm r}$', lw=2)
plt.semilogx(avals, ox0*avals**(-3*(1+wx))/rho_c_scaled(avals), color='c', label=r'$\rho_{\rm X}$', lw=2)
plt.xlabel('scale factor $a$')
plt.ylabel(r'$\Omega_{i}(a)$')
plt.legend(frameon=False)
plt.xlim([avals[0], avals[-1]])
show()
```

(a) The continuity equation gives us the evolution of the density, as derived in class:
\begin{equation}
\rho_i(a)=\rho_{0,i}\,a^{-3(1+w_i)}\,.
\end{equation}
If $w<-1$, then $-3(1+w) > 0$ and the density grows as the Universe expands. 

The density parameter $\Omega_{\rm X}(a)$ is defined as usual: 
\begin{equation*}
\Omega_{\rm X}(a)=\frac{\rho_{\rm X}(a)}{\rho_{\rm c}(a)}\,.
\label{eq:omega}\tag{6}
\end{equation*}
For this problem you're given that $k=0$ (flat Universe), so $\rho_{\rm c}(a)=\rho_{\rm X}(a)+\rho_{\rm m}(a)$. We can then do our usual trick of expressing $\rho_i(a)$ in terms of its present day value $\rho_{i,0}$ and its evolution with $a$:
$$ \rho_{\rm c}(a)=\rho_{\rm X,0}\,a^{-3(1+w_i)}+\rho_{\rm m,0}\,a^{-3}\;.$$ Going back to Equation \ref{eq:omega}, we get 
$$ \Omega_{\rm X}(a)=\frac{\rho_{\rm X,0}\,a^{-3(1+w_i)}}{\rho_{\rm X,0}\,a^{-3(1+w_i)}+\rho_{\rm m,0}\,a^{-3}} \;.$$
We can pull $\rho_{\rm X,0}\,a^{-3(1+w_{\rm X})}$ out of each term (in both the numerator and denominator) and divide every term by $\rho_{\rm c,0}$ (to get from $\rho \rightarrow \Omega$), which gives us 
$$ \Omega_{\rm X}(a)=\frac{1}{\frac{\Omega_{\rm m,0}}{\Omega_{\rm X,0}}\,a^{3\,w_i}+1}\,,$$
as desired.

If $\Omega_{\rm X,0}=0.75$ and $w_{\rm X}=-2$, we get 
$$ \Omega_{\rm X}(a_\star)=0.999 \rightarrow 0.999=\frac{1}{1+\frac{0.25}{0.75}a_\star^{-6}}\,$$
which gives $1/0.999=1+a_\star^{-6}/3$, so $a_\star^{-6}=0.003$ or $$a_\star=0.003^{-1/6} \approx 2.63\;.$$ 

(b) We are told that dark energy dominates the energy density at some time $t_\star$ (the value in part (a) satisifies this condition, thus my use of $a_\star$ there). If dark energy dominates the energy density, then it also dominates the contribution to $H(a_\star)$ since $H^2(a)=8\,\pi\,{\rm G}\,\sum_i \rho_{i}(a)/3$. We can therefore write 
\begin{equation*}
H(a)=H(a_\star)\left(\frac{a}{a_{\star}}\right)^{-3(1+w_X)/2}\,.
\label{eq:hubz_x} \tag{7}
\end{equation*}
(Note: we typically normalize everything at $a_0=1$, but we are now normalizing at $a_\star \ne 1$, thus the explicit dependence of $a_\star$ and $H(a_\star)$ in the equation.) Since we also have that $H(a)=\dot{a}/a$, we have a differential equation that we can integrate to find time:
\begin{equation*}
\frac{da}{dt}\frac{1}{a}=H(a_\star)\left(\frac{a}{a_{\star}}\right)^{-3(1+w_X)/2}\,.
\label{eq:hubz_diffeq} \tag{8}
\end{equation*}
Rearranging gives us 
$$ dt=\frac{da}{a_{\star}}\,H(a_\star)^{-1}\,\left(\frac{a}{a_{\star}}\right)^{3(1+w_X)/2-1}\,.$$
And now, we integrate:

$$ 
\begin{align}
\int_{t_\star}^{t_f} \,dt &=\int_{a_\star}^{a_f}\,\frac{da}{a_{\star}}\,H(a_\star)^{-1}\,\left(\frac{a}{a_{\star}}\right)^{3(1+w_X)/2-1}\,\\
H(a_\star)\,(t_f-t_\star) & = \frac{2}{3(1+w_X)}\,\left(\frac{a_f}{a_\star}\right)^{3(1+w_X)/2} \;\;(w \ne -1)\,.
\end{align}
$$
If $w<-1$, this integral converges as $a_f \rightarrow \infty$:
$$
H(a_\star)\,(t_f-t_\star)=-\frac{2}{3(1+w_X)}\,.
$$
Since the RHS of the equation is finite, so is the LHS, and we see that the Universe would reach infinite size ($a=\infty$) in a finite time in the future,  $$\Delta\,t=-\frac{2}{3\,(1+w_X)}\frac{1}{H_\star}$$ (recall $w_X < -1$, so $\Delta t > 0$).

(c) As we approach $t_f$ (the time of the Big Rip), we have $a \rightarrow \infty$. This means the wavelengths of CMB photons will be stretched further and further â€” $\lambda(a) = \lambda(a_{\star})\,(a/a_\star)=\lambda_0\,a$ â€”  approaching infinite wavelength (zero energy) and unobservability as we approach the Big Rip. We'd have to say good bye to the CMB, though that might not be highest on our list of worries at that point.
